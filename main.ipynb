```python
# 1. Memeriksa versi Python yang sedang digunakan
import sys
print(f"Python Version: {sys.version}")

# 2. Memeriksa apakah beberapa paket yang diinstal bekerja dengan baik
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 3. Membuat array sederhana dengan numpy dan plot menggunakan matplotlib
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y)
plt.title('Sine Wave Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.show()

# 4. Membaca DataFrame menggunakan pandas
data = {'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
        'Age': [24, 27, 22, 32, 29]}

df = pd.DataFrame(data)
print("Pandas DataFrame:")
print(df)

# 5. Menggunakan scikit-learn untuk membuat model sederhana
from sklearn.linear_model import LinearRegression

# Membuat data untuk regresi linear sederhana
X = np.array([[1], [2], [3], [4], [5]])
y = np.array([1, 2, 3, 4, 5])

model = LinearRegression()
model.fit(X, y)

print("Model Coefficients:", model.coef_)
print("Model Intercept:", model.intercept_)

# Menandakan bahwa environment berhasil
print("Environment `datascience_package` berhasil diaktifkan dan bekerja dengan baik!")
```
